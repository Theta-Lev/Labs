<h2>1. Аннотация</h2>
<hr>
<br>
<p style="text-align:justify">
    Исследуется энергетическая зависимость вероятности рассеяния электронов атомами инертного газа, определяются энергии электронов, при которых наблюдается «просветление» газа, и оценивается размер внешней электронной оболочки его атомов.
<p>

<h2>2. Теоретическое введение</h2>
<hr>
<br>
<p style="text-align:justify">
    К. Рамзауэр исследовал зависимость поперечных сечений упрогого рассеяния электронов (с энергией до 10 эВ) на атомах аргона. В результате этих исследований было обнаружено явление, получившее название <em>эффекта Рамзауэра</em>.
</p>
<p style="text-align:justify">
    Будем считать, что электрон рассеивается на одномерной прямоугольной потенциальной
яме конечной глубины. Решение задачи о прохождении частицы с энергией $E$ над потенциальной ямой шириной $l$ и глубиной $U_0$ можно найти из уравнения Шрёдингера. Коэффициент прохождения равен отношению квадратов амплитуд прошедшей и падающей волн и определяется выражением:
</p>
<p style="text-align:center">
    \begin{equation} \frac{1}{D} = 1 + \frac{{U_0}^2}{4E(E + U_0)} \sin^2(k_2l) \end{equation}
</p>
<p style="text-align:justify">
    Условие первого интерференционного максимума:
</p>
<p style="text-align:center">
    \begin{equation} 2l = \frac{h}{\sqrt{2m(E_2+U_0)}} \end{equation}
</p>
<p style="text-align:justify">
    Условие первого интерференционного минимума:
</p>
<p style="text-align:center">
    \begin{equation} 2l = \frac{3}{2} \frac{h}{\sqrt{2m(E_2+U_0)}} \end{equation}
</p>
<p style="text-align:justify">
    Измерив величины $E_1$ и $E_2$, можем рассчитать эффективную глубину потенциальной ямы атома:
</p>
<p style="text-align:center">
    \begin{equation} U_0 = \frac{4}{5} E_2 - \frac{9}{5} E_1 \end{equation}
</p>
<p style="text-align:justify">
    Тогда зависимость вероятности рассеяния электрона от его энергии можно определить следующим образом:
</p>
<p style="text-align:center">
    \begin{equation} w(U) = - \frac{1}{C} \ln \frac{I(U)}{I_0}, \end{equation}
</p>
<p style="text-align:justify">
    где $I_0$ -- ток катода, а $C$ -- некоторая постоянная
</p>

<h3>4.3 Обработка результатов</h3>

<p style="text-align:justify">
    По данным табл.1 оценена глубина потенциальной ямы
</p>
<p style="text-align:center">
    \begin{equation} U_0 = \frac{4}{5} E_2 - \frac{9}{5} E_1 = (3,0 \pm 0,7) В; \epsilon_{U_{0}} = 0,23 \end{equation}
</p>
<p style="text-align:justify">
    По данным табл.1 рассчитан размер электронной оболочки атома инертного газа в лампе (принимая $U_0 = (3,0 \pm 0,7) В$):
</p>
<p style="text-align:justify">
    Через первый интерференционный максимум
</p>
<p style="text-align:center">
    \begin{equation} l = \frac{h}{2\sqrt{2 m_e e (E_2+U_0)}} = (3,07 \pm 0,05) \times 10^{-10} м; \epsilon_{l} = 0,02 \end{equation}
</p>
<p style="text-align:justify">
    Через первый интерференционный минимум
</p>
<p style="text-align:center">
    \begin{equation} l = \frac{3}{4} \frac{h}{\sqrt{2 m_e e (E_2+U_0)}} = (3,07 \pm 0,13) \times 10^{-10} м; \epsilon_{l} = 0,04 \end{equation}
</p>
<p style="text-align:justify">
    Оценка размера без использования $U_0$
</p>
<p style="text-align:center">
    \begin{equation} l = \frac{\sqrt{7}}{4} \frac{h}{\sqrt{2 m_e e (E_2+E_1)}} = (3,06 \pm 0,14) \times 10^{-10} м; \epsilon_{l} = 0,05 \end{equation}
</p>
<p style="text-align:justify">
    Все оценки сходятся в пределах погрешностей. Это показывает, что применённая модель прямоугольной ямы достаточно точна.
</p>
<p style="text-align:justify">
    Построены графики ВАХ для статического режима (рис. 4 - 7), используя результаты таблиц 2 - 4.
</p>

```python
plt.figure(figsize=(16, 10))

plt.errorbar(raw_df1['V'], raw_df1['Va'], yerr = 0.2, xerr = 0.01, fmt = 'r.')
plt.title("Рис.4 ВАХ для Vнак = 2.98 В", fontsize=22)
plt.xlabel('V', fontsize = 18)
plt.ylabel('Va', fontsize = 18)
plt.show()
```

```python
plt.figure(figsize=(16, 10))

plt.errorbar(raw_df2['V'], raw_df2['Va'], yerr = 0.2, xerr = 0.01, fmt = 'g^')
plt.title("Рис.5 ВАХ для Vнак = 2.78 В", fontsize=22)
plt.xlabel('V', fontsize = 18)
plt.ylabel('Va', fontsize = 18)
plt.show()
```

```python
plt.figure(figsize=(16, 10))

plt.errorbar(raw_df3['V'], raw_df3['Va'], yerr = 0.2, xerr = 0.01, fmt = 'bs')
plt.title("Рис.6 ВАХ для Vнак = 2.58 В", fontsize=22)
plt.xlabel('V', fontsize = 18)
plt.ylabel('Va', fontsize = 18)
plt.show()
```

```python
plt.figure(figsize=(16,10))

plt.plot(raw_df1['V'], raw_df1['Va'], "ro",
         raw_df2['V'], raw_df2['Va'], "g^",
         raw_df3['V'], raw_df3['Va'], "bs")
plt.grid(visible=1, which='both', axis='both')
plt.title("Рис.7 Объединение ВАХ", fontsize=22)
plt.xlabel('V', fontsize = 18)
plt.ylabel('Va', fontsize = 18)
plt.show()
```

<p style="text-align:justify">
    Для статического режима тоже можно определить некоторые характеристики модели ($l$ -- размер внешней электронной оболочки, $U_0$ -- глубину потенциальной ямы). Так, для $V_{нак} = 2.98 В$:
</p>
<p style="text-align:center">
    \begin{equation}
    E_2 = (6,0 \pm 1,0) В; \epsilon_{E_{2}} = 0,17 \\ 
    E_1 = (1,0 \pm 0,5) В; \epsilon_{E_{1}} = 0,50 \\
    U_0 = (3,0 \pm 2,0) В; \epsilon_{U_{0}} = 0,57 \\
    l = (3,07 \pm 0,04) \times 10^{-10} м; \epsilon_{l} = 0,02
    \end{equation}
</p>
<p style="text-align:justify">
    Аналогично для других характеристик:
</p>
<p style="text-align:center">
    \begin{equation}
    U_0 = (3,8 \pm 1,7) В; \epsilon_{U_{0}} = 0,44 \\ 
    l = (2,87 \pm 0,05) В; \epsilon_{l} = 0,02 \\
    U_0 = (3,0 \pm 2,0) В; \epsilon_{U_{0}} = 0,57 \\
    l = (3,07 \pm 0,04) \times 10^{-10} м; \epsilon_{l} = 0,02
    \end{equation}
</p>
<p style="text-align:justify">
    Затем для $V_{нак} = 2.98 B$ построен график зависимости вероятности рассеяния энергии (рис.8)
</p>

```python
data_I = raw_df1['Va'] / 100.0e3
I_0 = raw_df1['V']
data_w = - np.log(data_I / I_0)
data_err = 0.2 / 100.0e3 * (1 / data_I + 1 / I_0)

plt.figure(figsize=(16, 10))

plt.errorbar(raw_df1['V'], data_w, yerr = data_err, xerr = 0.01, fmt = 'bs')
plt.title("Рис.8 Зависимость вероятности рассеяния энергии", fontsize=22)
plt.xlabel('V', fontsize = 18)
plt.ylabel('w(V)', fontsize = 18)
plt.show()
```
