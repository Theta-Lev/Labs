<h2>4. Метод обработки данных – метод наименьших квадратов для параболы</h2>
<hr>
<br>
<p style="text-align:justify">
    Так как в ходе работы получено много данных, которые необходимо будет проанализировать, то будет использован метод наименьших квадратов, который и будет сформулирован в этом блоке.
</p>
<p style="text-align:justify">
    Парабола выбрана как первое приближение модели "пиков" спектров; для этой модели функция имеет вид
</p>
<p style="text-align:center">
    \begin{equation}
        f(x) = ax^2 + bx + c;
    \end{equation}
</p>
<p style="text-align:justify">
    Метод тогда выглядит так
</p>
<p style="text-align:center">
    \begin{equation}
        F(a, b, c) = \sum\limits_{i = 1}^{n} \left(y_i - f(x_i)\right)^2;
    \end{equation}
</p>
<p style="text-align:justify">
    Найдем минимум этой функции, по необходимому условию
</p>
<p style="text-align:center">
    \begin{equation}
        \frac{\partial F}{\partial a} = \sum\limits_{i = 1}^n  -2(y_i - ax_i^2 - bx_i - c)(-x_i^2) = 0;
    \end{equation}
    \begin{equation}
        \frac{\partial F}{\partial b} = \sum\limits_{i = 1}^n  -2(y_i - ax_i^2 - bx_i - c)(-x_i) = 0;
    \end{equation}
    \begin{equation}
        \frac{\partial F}{\partial c} = \sum\limits_{i = 1}^n  -2(y_i - ax_i^2 - bx_i - c)(-1) = 0.
    \end{equation}
</p>
<p style="text-align:justify">
    Решив систему, получим
</p>
<p style="text-align:center">
    \begin{equation}
        a = \frac{\left<x_i^2y_i\right> - b \left<x_i^3\right> - c \left<x_i^2\right>}{\left<x_i^4\right>};
    \end{equation}
    \begin{equation}
        b = \frac{\left( \left<x_i^4\right> \left<x_iy_i\right> - \left<x_i^2\right> \left<x_i^2y_i\right> \right) - c \left( \left<x_i^4\right> \left<x_i\right> - \left<x_i^2\right> \right) } { \left<x_i^4\right> \left<x_i^2\right> - \left<x_i^3\right> };
    \end{equation}
    \begin{equation}
        c = \frac{ \left<x_i^4\right>^2 \left( \left<x_i^2\right> \left<y_i\right> - \left<x_i\right> \left<x_iy_i\right> \right) + \left<x_i^4\right> \left<x_i^3\right> \left( \left<x_iy_i\right> - \left<y_i\right> \right) + \left<x_i^4\right> \left<x_i^2y_i\right> \left( \left<x_i^3\right> \left<x_i\right> - \left<x_i^2\right>^2 \right) + \left<x_i^3\right> \left<x_i^2y_i\right> \left( \left<x_i^2\right> - \left<x_i^3\right> \right) } { \left<x_i^4\right> \left<x_i^2\right> \left( \left<x_i^4\right> - \left<x_i^2\right> \right) + \left<x_i^4\right> \left<x_i\right> \left( \left<x_i^3\right> - \left<x_i^4\right> \left<x_i\right> \right) + \left<x_i^4\right> \left( \left<x_i\right> \left<x_i^2\right> - \left<x_i^3\right> \right)}.
    \end{equation}
</p>
<p style="text-align:justify">
    Находим пик как
</p>
<p style="text-align:center">
    \begin{equation}
        x_c = \frac{-b}{2a};
    \end{equation}
</p>
<p style="text-align:justify">
    Погрешность будет равна
</p>
<p style="text-align:center">
    \begin{equation}
        \sigma_{x_c} = x_c\sqrt{\left(\frac{\sigma_b}{b}\right)^2 + \left(\frac{\sigma_a}{a}\right)^2};
    \end{equation}
</p>
<br>
<hr>
<br>
<p style="text-align:justify">
    В некоторых опытах нам может понадобится по графику определить канал переломной точки. Это мы можем сделать, построив две прямые по мнк по формуле для прямой: $y = kx + b$, тогда из МНК
</p>
<p style="text-align:center">
    \begin{equation}
        k = \frac{\left<xy\right>  - \left<x\right>\left<y\right>}{\left<x^2\right> - \left<x\right>^2};
    \end{equation}
    \begin{equation}
        \sigma_k = \sqrt{\frac{1}{n-1} \left(\frac{\left<y^2\right>}{\left<y\right>^2} - k^2\right)};
    \end{equation}
    \begin{equation}
        b = (\left<y\right> - k\left<x\right>);
    \end{equation}
    \begin{equation}
        \sigma_b =  \sigma_k\sqrt{\left<x^2\right>}.
    \end{equation}
</p>
<p style="text-align:justify">
    После аппроксимации у нас будут две прямые: $f_1(x) = k_1 x + b_1$,$f_2(x) = k_2 x + b_2$. Найдем их точку пересечения. Будем все делать только для $x$, поскольку нас интересует только номер канала
</p>
<p style="text-align:center">
    \begin{equation}
        x = \frac{b_2 - b_1}{k_1 - k_2};
    \end{equation}
    \begin{equation}
        \sigma_x = \sqrt{\left( \frac{\sigma_{b_1}}{b_1}\right)^2 + \left( \frac{\sigma_{b_2}}{b_2}\right)^2 + \left( \frac{\sigma_{k_1}}{k_1}\right)^2 + \left( \frac{\sigma_{k_2}}{k_2}\right)^2}.
    \end{equation}
</p>

<h2>5. Программная часть</h2>
<hr>
<br>
<p style="text-align:justify">
    В работе будет анализироваться по одному и тому же шаблону большое количество данных, которые собраны в одном и том же виде, то, для уменьшения вероятности ошибок в следюущем блоке кода будут собраны функции, которые потом будут вызываться при анализе 
</p>

```python
# Подключение необходимых библиотек

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Реализация вывода данных из памяти

def draw_raw(df, TITLE):
    plt.figure(figsize=(16, 10))

    plt.errorbar(df['Channel'], df['Counts'], fmt = 'x')
    plt.title(TITLE, fontsize=22)
    plt.xlabel('Channel', fontsize = 18)
    plt.ylabel('Counts', fontsize = 18)
    plt.show()

# Реализация МНК для параболы

def P(a, b, c, x):
    return a * x * x + b * x + c

def parabola(df):
    dfx = list(df['Channel'])
    dfy = list(df['Counts'])
    
    # dx = lambda x: 1
    # dy = lambda y: 0.01*y
    
    ax1 =    sum(map(lambda x: x**1, dfx))/len(dfx)
    dx1 =    max((list(map(lambda x: 1, dfx)))/np.sqrt(len(dfx)))
    ax2 =    sum(map(lambda x: x**2, dfx))/len(dfx)
    dx2 =    max((list(map(lambda x: 2, dfx)))/np.sqrt(len(dfx)))
    ax3 =    sum(map(lambda x: x**3, dfx))/len(dfx)
    dx3 =    max((list(map(lambda x: 3, dfx)))/np.sqrt(len(dfx)))
    ax4 =    sum(map(lambda x: x**4, dfx))/len(dfx)
    dx4 =    max((list(map(lambda x: 4, dfx)))/np.sqrt(len(dfx)))
    ay1 =    sum(map(lambda y: y**1, dfy))/len(dfx)
    dy1 =    max((list(map(lambda y: 0.01*y, dfy)))/np.sqrt(len(dfx)))
    ax1y1 =  sum(map(lambda x, y: x*y, dfx, dfy))/len(dfx)
    dx1y1 =  max((list(map(lambda x, y: 1 + 0.01*y, dfx, dfy)))/np.sqrt(len(dfx)))
    ax2y1 =  sum(map(lambda x, y: x*x*y, dfx, dfy))/len(dfx)
    dx2y1 =  max((list(map(lambda x, y: 2 + 0.01*y, dfx, dfy)))/np.sqrt(len(dfx)))
    
    #print(ax1, ax2, ax3, ax4, ay1, ax1y1, ax2y1)
    #print(dx1, dx2, dx3, dx4, dy1, dx1y1, dx2y1)
    
    c = (
            ((ax4)**2)*((ax2)*(ay1) - (ax1)*(ax1y1)) +
            ((ax4)*(ax3))*((ax1y1) - (ay1)) +
            ((ax4)*(ax2y1))*((ax3)*(ax1) - (ax2)**2) +
            ((ax3)*(ax2y1))*((ax2) - (ax3))
        ) / (
            ((ax4)*(ax2))*((ax4) - (ax2)) +
            ((ax4)*(ax1))*((ax3) - (ax4)*(ax1)) +
            (ax4)*((ax2)*(ax1) - (ax3))
        )
    dc = c * np.sqrt(
                        ((ax2)**2)*(((ay1)**2) + 1)*((2*dx4/ax4)**2) +
                        ((ax2y1)**2)*((2*dx3/ax3)**2) +
                        (((ax4)*(ax2y1))**2)*((2*dx2/ax2)**2) +
                        ((ax4)**2)*((2*dx1/ax1)**2)
                    )
    
    b = (
            (ax1y1)*(ax4) -
            (ax2y1)*(ax3) + 
            c * ( (ax2) - (ax1)*(ax4) )
        ) / (
            (ax4)*(ax2) - (ax3)
        )
    db = b * np.sqrt(
                        ((c*(ax1))**2 + (ax2)**2)*((dx1/ax1)**2) +
                        ((ax2y1)**2 + (1)**2)*((dx3/ax3)**2) +
                        ((c)**2 + (ax4)**2)*((dx2/ax2)**2) +
                        ((c*(ax4))**2)*((dx1/ax1)**2) + 
                        (((ax4)*(ax1) - (ax2))**2)*((dc/c)**2)
                    )
    
    a = (
            (ax2y1) -
            b * (ax3) -
            c * (ax2)
        ) / (
            (ax4)
        )
    da = a * np.sqrt(
                        ((dx4/ax4)**2) +
                        (b**2)*((dx3/ax3)**2) +
                        (c**2)*((dx2/ax2)**2) +
                        ((ax3)**2)*((db/b)**2) +
                        ((ax2)**2)*((dc/c)**2)
                    )
    #print(a, b, c)
    #print(da, db, dc)
    return a, b, c, da, db, dc

# Функция поиска вершины параболы

def vertex(a, b, da, db):
    x = -1 * float(b)/float(2*a)
    dx = x * np.sqrt(
                        1*(1*da/a)**2 +
                        1*(1*db/b)**2
                    )
    x = int(x)
    dx = int(dx)
    
    return x, dx

# Реализация МНК для прямой

def L(k, b, x):
    return k * x + b

def line(df):
    dfx = list(df['Channel'])
    dfy = list(df['Counts'])
    
    # dx = lambda x: 1
    # dy = lambda y: 0.01*y
    
    ax1 =    sum(map(lambda x: x**1, dfx))/len(dfx)
    dx1 =    max((list(map(lambda x: 1, dfx))/np.sqrt(len(dfx))))
    ax2 =    sum(map(lambda x: x**2, dfx))/len(dfx)
    dx2 =    max((list(map(lambda x: 2, dfx))/np.sqrt(len(dfx))))
    ay1 =    sum(map(lambda y: y**1, dfy))/len(dfx)
    dy1 =    max((list(map(lambda y: 0.01*y, dfy))/np.sqrt(len(dfx))))
    ay2 =    sum(map(lambda y: y**2, dfy))/len(dfx)
    dy2 =    max((list(map(lambda y: 0.02*y, dfy))/np.sqrt(len(dfx))))
    ax1y1 =  sum(map(lambda x, y: x*y, dfx, dfy))/len(dfx)
    dx1y1 =  max((list(map(lambda x, y: 1 + 0.01*y, dfx, dfy))/np.sqrt(len(dfx))))
    
    k = (
            (ax1y1) - (ax1)*(ay1)
        ) / (
            (ax2) - (ax1)**2
        )
    dk = np.sqrt( ((ay2)/(ay1)**2 - k**2)/(len(df) - 1) )
    
    b = (ay1) - k*(ax1)
    db = dk * np.sqrt( (ax2) )
    
    return k, b, dk, db

# Функция поиска пересечения двух прямых

def point(k1, b1, k2, b2, dk1, db1, dk2, db2):
    x = (b2 - b1)/(k1 - k2)
    dx = x * np.sqrt(
                        (dk1/k1)**2 + (db1/b1)**2 +
                        (dk2/k2)**2 + (db2/b2)**2
                    )
    x = int(x)
    
    return x, dx

# Отрисовка итоговых графиков

def draw_par(df, a, b, c, x, dx, l, TITLE):
    data_err = 0.01 * df['Counts']
    
    x = int(x)
    x_par = np.array([i for i in range(x-l, x+l)])
    y_par = P(a, b, c, x_par)
    
    x_pt = np.array([x])
    y_pt = P(a, b, c, x_pt)
    
    plt.figure(figsize=(16, 10))

    plt.errorbar(df['Channel'], df['Counts'], yerr = data_err, xerr = 1, fmt = '.')
    plt.plot(x_par, y_par, color='g', label = 'Аппроксимирующая парабола')
    plt.scatter(x_pt, y_pt, color='r')
    plt.errorbar(x_pt, y_pt, xerr=1, fmt="o", color="r")
    plt.title(TITLE, fontsize=22)
    plt.xlabel('Channel', fontsize = 18)
    plt.ylabel('Counts', fontsize = 18)
    plt.show()
    
def draw_lin(df, k1, b1, k2, b2, x, dx, l, TITLE):
    data_err = 0.01 * df['Counts']
    
    x = int(x)
    x_lin = np.array([i for i in range(x-l, x+l+1)])
    y1_lin = L(k1, b1, x_lin)
    y2_lin = L(k2, b2, x_lin)
    
    x_pt = np.array([x])
    y_pt = L(k1, b1, x_pt)
    
    plt.figure(figsize=(16, 10))

    plt.errorbar(df['Channel'], df['Counts'], yerr = data_err, xerr = 1, fmt = '.')
    plt.plot(x_lin, y1_lin, color='g', label = 'Аппроксимирующая прямая')
    plt.plot(x_lin, y2_lin, color='g', label = 'Аппроксимирующая прямая')
    plt.scatter(x_pt, y_pt, color='r')
    plt.errorbar(x_pt, y_pt, xerr=1, fmt="o", color="r")
    plt.title(TITLE, fontsize=22)
    plt.xlabel('Channel', fontsize = 18)
    plt.ylabel('Counts', fontsize = 18)
    plt.show()
```
